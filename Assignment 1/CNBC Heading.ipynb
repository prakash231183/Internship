### Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world :
#### i) Headline
#### iii) News Link

```python
# import importent libraries and get the response of request

import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
url = 'https://www.cnbc.com/world/?region=world'
r = requests.get(url)
r
```

```python
soup=bs(r.content)
```

```python
titles=[]

for i in soup.find_all('a',class_="LatestNews-headline"):
    titles.append(i.text)
    
    
time=[]

for i in soup.find_all('time',class_="LatestNews-timestamp"):
    time.append(i.text)
    
    
url=[]

for i in soup.find_all('a',class_="LatestNews-headline"):
    url.append(i.get('href'))   
```

### Making a Data Frame

```python
import pandas as pd
df=pd.DataFrame({'Title':titles,'Time':time,'Url':url})
df
```
