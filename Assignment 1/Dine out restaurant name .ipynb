### Write a python program to scrape mentioned details from dineout.co.in :
#### i) Restaurant name
#### ii) Cuisine
#### iii) Location
#### iv) Ratings
#### v) Image URL

```python
# import important libraries

import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
url = 'https://www.dineout.co.in/delhi-restaurants/buffet-special'
r = requests.get(url)
r
```

```python
# scraping Data 

soup=bs(r.content)

titles=[]
cuisin=[]
location=[]
rating=[]
url=[]

for i in soup.find_all('a',class_="restnt-name ellipsis"):
    titles.append(i.text)
    
    
for i in soup.find_all('span',class_="double-line-ellipsis"):
    cuisin.append(i.text.split('|')[1])
    
    
for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    location.append(i.text)
    

for i in soup.find_all('div',class_="restnt-rating rating-4"):
    rating.append(i.text)
    

for i in soup.find_all('img',class_="no-img"):
    url.append(i.get('data-src'))
    
```

### Making DataFrame

```python
import pandas as pd
df=pd.DataFrame({'Title':titles,'Location':location,'Rating':rating,'Cusins':cuisin,'Image url':url})
df
```
